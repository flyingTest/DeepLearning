# 支持向量机(SVM)<p>

[算法原理见此链接](https://blog.csdn.net/qq_41664845/article/details/82790990)<p>
与之前相同。本节依然使用iris数据集来创建SVM的线性分类器。<p>
来预测数据是否为山鸢尾花<p>
山鸢尾为1，其他种类花为-1<p>
另外这里我们只取2个特征，这是为了方便之后的可视化。作为一个二维的图像展示

```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from sklearn import datasets
#重新初始化tf
from tensorflow.python.framework import ops
ops.reset_default_graph()

sess = tf.Session()

#读数据
iris = datasets.load_iris()
x = np.array([[x[0],x[3]] for x in iris.data])
y = np.array([1. if x==0 else -1. for x in iris.target ])
```

* 分割数据集

将数据分为80%训练集和20%测试集

```python
train_data_index = np.random.choice(len(x),round(len(x)*0.8),replace=False)
test_data_index = np.array(list(set(range(len(x))) - set(train_data_index)))

x_train = x[train_data_index]
x_test = x[test_data_index]
y_train = y[train_data_index]
y_test = y[test_data_index]
```

* 声明批量大小，占位符，变量和模型变量

```python
batch_size = 100

x_data = tf.placeholder(shape=[None,2],dtype=tf.float32)
y_target = tf.placeholder(shape=[None,1],dtype=tf.float32)

w = tf.Variable(tf.random_normal(shape=[2,1]))
b = tf.Variable(tf.random_normal(shape=[1,1]))
```

```python
output = tf.subtract(tf.matmul(x_data,w),b)
```

### SVM的soft margin 损失函数公式：$ \frac{1}{n}\sum_{i=1}^{n}max(0,1-y_{i}(wx_{i}-b))+a\left \| w \right \|^2 $
其中W是斜率向量，b是截距，$x_{i}$是输入向量，$y_{i}$是实际分类（-1或者1），a是软分类器的正则参数

```python
L2_norm = tf.reduce_sum(tf.square(w))

alpha = tf.constant([0.1])

classification_term = tf.reduce_mean(tf.maximum(0.,tf.subtract(1.,tf.multiply(output,y_target))))

loss = tf.add(classification_term,tf.multiply(alpha,L2_norm))
```

* 计算准确度

```python
prediction = tf.sign(output)
prediction_correct = tf.cast(tf.equal(prediction,y_target),tf.float32)
accuracy = tf.reduce_mean(prediction_correct) 
```

* 声明优化器

```python
opt = tf.train.GradientDescentOptimizer(0.01)
train_step = opt.minimize(loss)
```

```python
init = tf.global_variables_initializer()
sess.run(init)
```

```python
epoch=500
train_loss_result = []
test_loss_result = []
train_acc_result = []
test_acc_result = []

for i in range(epoch):
    rand_index = np.random.choice(len(x_train),size=batch_size)
    rand_x = x_train[rand_index]
    rand_y = np.transpose([y_train[rand_index]])
    sess.run(train_step,feed_dict={x_data:rand_x,y_target:rand_y})
   
    train_loss = sess.run(loss,feed_dict={x_data:rand_x,y_target:rand_y})
    train_loss_result.append(train_loss)
    
    train_acc = sess.run(accuracy,feed_dict={x_data:rand_x,y_target:rand_y})
    train_acc_result.append(train_acc)    
    
    test_loss = sess.run(loss,feed_dict={x_data:x_test,y_target:np.transpose([y_test])})
    test_loss_result.append(test_loss)
    
    test_acc = sess.run(accuracy,feed_dict={x_data:x_test,y_target:np.transpose([y_test])})
    test_acc_result.append(test_acc)  
    
    
    if (i+1)% 100 ==0:
        print('epoch:{},train loss：{}，train acc：{}，test loss：{}，test acc：{}'.format(i+1,train_loss,train_acc,test_loss,test_acc))
```

## 可视化结果

```python
[[feature1],[feature2]] = sess.run(w)
[_b] = sess.run(b)
slope = -feature2/feature1
y_intercept = _b/feature1

x1 = [i[1] for i in x]

best_fit = []
for i in x:
    best_fit.append(slope*i+y_intercept)
    
setosa_x = [v[1] for i,v in enumerate(x) if y[i]==1]
setosa_y = [v[0] for i,v in enumerate(x) if y[i]==1]
not_setosa_x = [v[1] for i,v in enumerate(x) if y[i]==-1]
not_setosa_y = [v[1] for i,v in enumerate(x) if y[i]==-1]
```

```python
plt.plot(setosa_x,setosa_y,'o',label='I.setosa')
plt.plot(not_setosa_x,not_setosa_x,'x',label='Non-setosa')
plt.plot(x1,best_fit,'r-',label='Liner Separator',linewidth=3)
plt.ylim([0,10])
plt.title('sepal length vs pedal width')
plt.xlabel('pedal width')
plt.ylabel('sepal length')
```

## 结果分析
使用TensorFlow实现SVD算法可能导致每次运算的结果都不尽相同。<p>原因包括训练集和测试集的随机分批，每次训练的批量大小不同。在理想情况下每次迭代后学习率缓慢减小。

```python
plt.plot(train_acc_result,'k-',label='train accuracy')
plt.plot(test_acc_result,'r--',label='test accuracy')
plt.legend(loc='upper right')
plt.title('train and test set accuracies')
plt.ylabel('accuracy')
plt.xlabel('generation')
```

```python
plt.plot(train_loss_result,'k-',label='train loss')
plt.plot(test_loss_result,'r--',label='test loss')
plt.legend(loc='upper right')
plt.title('loss per generation')
plt.ylabel('loss')
plt.xlabel('generation')
```

# TensorFlow 上核函数的使用
上面介绍了SVM处理线性可分数据集，那么如何处理非线性数据集呢？<p>
下面为大家介绍一下SVM损失函数中的核函数<p>
上一节我们采用特殊的损失函数来解决soft margin问题。另外一种解决soft margin的方法是对偶优化问题。<p>
$ max\sum_{i=1}^{n}b_{i}-\frac{1}{2}\sum_{i=1}^{n}\sum_{j=1}^{n}y_{i}y_{j}b_{i}b_{j}(x_{i}*x_{j}) $<p>
其中：<p>
$ \sum_{i=1}^{n}b_{i}y_{i}=0 $  和  $ 0 \leq b_{i}\geq \frac{1}{2ny} $  

```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from sklearn import datasets
#重新初始化tf
from tensorflow.python.framework import ops
ops.reset_default_graph()

sess = tf.Session()
```

生成同心圆的模拟数据，之后再进行筛选确保只有类-1或者1

```python
(x,y) = datasets.make_circles(n_samples=500,factor=.5,noise=.1)
y = np.array([1 if i==1 else -1 for i in y])
class1_x = [value[0] for i,value in enumerate(x) if y[i] ==1]
class1_y = [value[1] for i,value in enumerate(x) if y[i] ==1]
class2_x = [value[0] for i,value in enumerate(x) if y[i] ==-1]
class2_y = [value[1] for i,value in enumerate(x) if y[i] ==-1]
```

```python
batch_size = 500
x_data = tf.placeholder(shape=[None,2],dtype=tf.float32)
y_target = tf.placeholder(shape=[None,1],dtype=tf.float32)

prediction_grid = tf.placeholder(shape=[None,2],dtype=tf.float32)
b = tf.Variable(tf.random_normal(shape=[1,batch_size]))
```

创建高斯核函数 (也称为径向基核函数或者RBF和函数)，公式如下：<p>
### $ k(x_{i},x_{j}) = e^{-\gamma \left \| x_{i}-x_{i} \right \|^2} $
   
### $ x\rightarrow (e^{-\gamma \left \| x_{i}-l_{1} \right \|^2},e^{-\gamma \left \| x_{i}-l_{2} \right \|^2}) $

```python
gamma = tf.constant(-44.0)
dist = tf.reduce_sum(tf.square(x_data), 1)
dist = tf.reshape(dist,[-1,1])
sq_dists = tf.add(tf.subtract(dist,tf.multiply(2.,tf.matmul(x_data,tf.transpose(x_data)))),tf.transpose(dist))
kernel = tf.exp(tf.multiply(gamma,tf.abs(sq_dists)))
```

本节提到的对偶问题是为了最大化，所以这里我们对loss取负。<p>
对偶优化问题。<p>
$ max\sum_{i=1}^{n}b_{i}-\frac{1}{2}\sum_{i=1}^{n}\sum_{j=1}^{n}y_{i}y_{j}b_{i}b_{j}(x_{i}*x_{j}) $<p>
其中：<p>
$ \sum_{i=1}^{n}b_{i}y_{i}=0 $  和  $ 0 \leq b_{i}\geq \frac{1}{2ny} $  <p>
    由此构建损失函数

```python
output = tf.matmul(b,kernel)
first_term = tf.reduce_sum(b)
b_vec_cross = tf.matmul(tf.transpose(b),b)
y_target_cross = tf.matmul(y_target,tf.transpose(y_target))
second_term = tf.reduce_sum(tf.multiply(kernel,tf.multiply(b_vec_cross,y_target_cross)))
loss = tf.negative(tf.subtract(first_term,second_term))
```

```python
r_A = tf.reshape(tf.reduce_sum(tf.square(x_data),1),[-1,1])
r_B = tf.reshape(tf.reduce_sum(tf.square(prediction_grid),1),[-1,1])
pred_sq_dist = tf.add(tf.subtract(r_A,tf.multiply(2.,tf.matmul(x_data,tf.transpose(prediction_grid)))),tf.transpose(r_B))
pred_kernel = tf.exp(tf.multiply(gamma,tf.abs(pred_sq_dist)))

prediction_output = tf.matmul(tf.multiply(tf.transpose(y_target),b),pred_kernel)
prediction = tf.sign(prediction_output-tf.reduce_mean(prediction_output))
accuracy = tf.reduce_mean(tf.cast(tf.equal(tf.squeeze(prediction),tf.squeeze(y_target)),tf.float32))
```

* 定义优化器

```python
opt = tf.train.GradientDescentOptimizer(0.001)
train_step = opt.minimize(loss)
```

```python
init = tf.global_variables_initializer()
sess.run(init)
```

* 开始训练，使用随机批量数据

```python
epoch=1500
train_loss_result = []
train_acc_result = []

for i in range(epoch):
    rand_index = np.random.choice(len(x),size=batch_size)
    rand_x = x[rand_index]
    rand_y = np.transpose([y[rand_index]])
    sess.run(train_step,feed_dict={x_data:rand_x,y_target:rand_y})
   
    train_loss = sess.run(loss,feed_dict={x_data:rand_x,y_target:rand_y})
    train_loss_result.append(train_loss)
    
    train_acc = sess.run(accuracy,feed_dict={x_data:rand_x,y_target:rand_y,prediction_grid:rand_x})
    train_acc_result.append(train_acc)    

    
    
    if (i+1)% 500 ==0:
        print('epoch:{},train loss：{}，train acc：{}'.format(i+1,train_loss,train_acc))
```

## 可视化

```python
x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
y_min, y_max = x[:, 1].min() - 1, x[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange( x_min, x_max,0.02),
                     np.arange( y_min, y_max,0.02))
grid_points = np.c_[xx.ravel(), yy.ravel()]

[grid_predictions] = sess.run(prediction,feed_dict={x_data:rand_x,y_target:rand_y,prediction_grid:grid_points})
grid_predictions = grid_predictions.reshape(xx.shape)
```

```python
plt.contourf(xx,yy, grid_predictions, camp=plt.cm.Paired, alpha=0.8)
plt.plot(class1_x,class1_y,'ro',label='class 1')
plt.plot(class2_x,class2_y,'kx',label='class 2')
plt.legend(loc='lower right')
plt.ylim([-1.5,1.5])
plt.xlim([-1.5,1.5])
```

你可以在这里改变gamma值来观察分类图像的变化。<p>
gamma的值越大，每个数据点对分类边界的影响就越大

```python
plt.plot(train_acc_result,'k-',label='Accuracy')
plt.title('batch accuracy')
plt.xlabel('generation'),plt.ylabel('Accuracy')
plt.legend(loc='lower right')
```
