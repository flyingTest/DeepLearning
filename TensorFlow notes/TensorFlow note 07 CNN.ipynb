# 使用TensorFolow实现简单的CNN 
本节将开发一个四层的卷积神经网络模型，提升预测MNIST数字的准确度<p>
前2个卷积层由Convolution-Relu-maxpool操作组成，后面2层是全连接层

```python
import tensorflow as tf
import numpy as np
from tensorflow.examples.tutorials.mnist import input_data
import time
```

首先 加载数据集，转化为图像28*28的数组

```python
data_dir = 'temp'
mnist = input_data.read_data_sets(data_dir,one_hot=True)
```

config 参数设置

```python
epoch = 5
batch_size= 50
n_batch=mnist.train.num_examples//batch_size
train_size = mnist.train.images.shape[0]
image_size = mnist.train.images.shape[1]
target_size = mnist.train.labels.shape[1]
image_hight = int(np.sqrt(image_size))
image_width = int(np.sqrt(image_size))
channels = 1
max_pool_size = 2
cov1_feature = 32
cov2_feature = 64

fully1_node = 84

h_pool2_hight,h_pool2_width = 7,7
```

1. 定义初始化权值函数
2. 定义初始化偏置函数
3. 定义初始化卷积层
4. 定义初始化池化层

```python
def weight_variable(shape):
    initial=tf.truncated_normal(shape,stddev=0.1)
    return tf.Variable(initial)

def bias_variable(shape):
    initial=tf.constant(0.1,shape=shape)
    return tf.Variable(initial)

def conv2d(input,filter):
    return tf.nn.conv2d(input,filter,strides=[1,1,1,1],padding='SAME')

def max_pool_2x2(value):
    return tf.nn.max_pool(value,ksize=[1,max_pool_size,max_pool_size,1],strides=[1,max_pool_size,max_pool_size,1],padding='SAME')
```

1. 输入层，data_x和y_target的占位符,并将data_x转化为4维的图像形状 [batch,hight,width,channels]<p>
2. 对于2个卷积层:</p>
   * 卷积 首先我们初始化其权重和偏置，权值由采样窗口（filter），卷积核决定,偏置为卷积核数量</p>
   * 激励 relu激励层对卷积结果激活</p>
   * 池化 max_pooling，或者avg_pooling</p>


```python
#输入层
x=tf.placeholder(tf.float32,[None,image_size]) 
y=tf.placeholder(tf.float32,[None,target_size])
x_image=tf.reshape(x,[-1,image_hight,image_width,channels])
    
#卷积层I
W_conv1=weight_variable([5,5,channels,cov1_feature]) 
b_conv1=bias_variable([cov1_feature]) 
h_conv1=tf.nn.relu(tf.nn.bias_add(conv2d(x_image,W_conv1),b_conv1))
h_pool1=max_pool_2x2(h_conv1) 

#卷积层II
W_conv2=weight_variable([5,5,cov1_feature,cov2_feature])
b_conv2=bias_variable([cov2_feature])
h_conv2=tf.nn.relu(conv2d(h_pool1,W_conv2)+b_conv2)
h_pool2=max_pool_2x2(h_conv2)  
```

3. 对于2个全连接层
  * 初始化全连接层的权值  全连接层输入神经元计算与是否padding,maxpool,cov2_feature有关。
  * 将池化层输出扁平化处理
  * 再通过sigmoid或者relu激活函数,
  * 加入dropout增加模型的泛化能力,结果作为下一层的输入

```python

W_fc1=weight_variable([h_pool2_hight*h_pool2_width*cov2_feature,fully1_node])
b_fc1 = bias_variable([fully1_node])

h_pool2_flat = tf.reshape(h_pool2,[-1,h_pool2_hight*h_pool2_width*cov2_feature])
h_fc1=tf.nn.relu(tf.matmul(h_pool2_flat,W_fc1)+b_fc1)
 
keep_prob=tf.placeholder(tf.float32)
h_fc1_drop=tf.nn.dropout(h_fc1,keep_prob)
 
W_fc2=weight_variable([fully1_node,target_size])
b_fc2=bias_variable([target_size])
```

```python
output = tf.matmul(h_fc1_drop,W_fc2)+b_fc2
```

 本次训练不是二分类，所以对最后结果使用softmax作为损失函数

```python
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=output,labels=y))                
```

创建训练集和测试集的预测函数<p>
结果存放在一个布尔列表中(argmax函数返回一维张量中最大的值所在的位置)</p>
求准确率(tf.cast将布尔值转换为float型)

```python
prediction = tf.nn.softmax(output)

correct_prediction=tf.equal(tf.argmax(prediction,1),tf.argmax(y,1))
accuracy=tf.reduce_mean(tf.cast(correct_prediction,tf.float32))
```

优化器 使用AdamOptimizer进行优化

```python
train_step=tf.train.AdamOptimizer(1e-4).minimize(loss)
```

```python
train_loss_result,train_acc_result = [],[]
test_loss_result,test_acc_result = [],[]
with tf.Session() as sess:
    start_time=time.clock()
    sess.run(tf.global_variables_initializer()) 
    for i in range(epoch):  
        for batch in range(n_batch):
            batch_xs,batch_ys=mnist.train.next_batch(batch_size)
            sess.run(train_step,feed_dict={x:batch_xs,y:batch_ys,keep_prob:0.7}) 
            train_loss,train_acc = sess.run([loss,accuracy],feed_dict={x:batch_xs,y:batch_ys,keep_prob:0.7}) 
            test_loss,test_acc=sess.run([loss,accuracy],feed_dict={x:mnist.test.images,y:mnist.test.labels,keep_prob:1.0})
            train_loss_result.append(train_loss)
            train_acc_result.append(train_acc)
            test_loss_result.append(test_loss)
            test_acc_result.append(test_acc)
            reslut = [i+1,batch,train_loss,test_loss,train_acc,test_acc]
            resluts = [ np.round(x,2) for x in reslut]
            print('epoch:{},batch:{},train_loss(test_loss):{:.2f}({:.2f}),train_acc(test_acc):{:.2f}({:.2f})'.format(*resluts))
    end_time=time.clock()
    print('Running time:%s Second'%(end_time-start_time)) 
```

# 最后可视化Loss及ACC曲线

```python
import matplotlib.pyplot as plt
_x = [i for i in range(1,len(train_loss_result)+1)]
plt.plot(_x, train_loss_result, 'k-', label='Train Loss')
plt.plot(_x, test_loss_result, 'r--', label='Test Loss', linewidth=4)
plt.title('Cross Entropy Loss per Generation')
plt.xlabel('Generation')
plt.ylabel('Cross Entropy Loss')
plt.legend(loc='upper right')
plt.show()
plt.plot(_x, train_acc_result, 'k-', label='Train Set Accuracy')
plt.plot(_x, test_acc_result, 'r--', label='Test Set Accuracy', linewidth=4)
plt.title('Train and Test Accuracy')
plt.xlabel('Generation')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
```
