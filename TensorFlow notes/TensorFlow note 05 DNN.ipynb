### 使用TensorFlow门函数和激励函数(example: sigmod,ReLU)
$ sigmoid(x)=\frac{1}{1+e^{x}}$<p>$ ReLU(x)=max(0,x)$<p>

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.python.framework import ops
ops.reset_default_graph()
sess = tf.Session()
tf.set_random_seed(5)
np.random.seed(42)
```

现在我们创建2个相同结构的单层神经网络，有一点不同的是<p>
一个使用sigmoid，一个使用ReLU作为激活函数，损失函数使用L2范数距离函数<p>
输出结果与0.75的差值

```python
batch_size = 50
sig_w = tf.Variable(tf.random_normal(shape=[1,1]))
sig_b = tf.Variable(tf.random_uniform(shape=[1,1]))
relu_w = tf.Variable(tf.random_normal(shape=[1,1]))
relu_b = tf.Variable(tf.random_uniform(shape=[1,1]))

x = np.random.normal(2, 0.1, 500)
x_data = tf.placeholder(shape=[None, 1],dtype=tf.float32)
```

声明2个训练模型 一个使用sigmoid，一个使用relu作为激活函数。

```python
sigmoid_activation = tf.sigmoid(tf.add(tf.matmul(x_data,sig_w),sig_b))
relu_activation = tf.nn.relu(tf.add(tf.matmul(x_data,relu_w),relu_b))
```

计算激活函数后与0.75之前的差值的L2范数平均

```python
sig_loss = tf.reduce_mean(tf.square(tf.subtract(sigmoid_activation,0.75)))
relu_loss = tf.reduce_mean(tf.square(tf.subtract(relu_activation,0.75)))
```

```python
opt = tf.train.GradientDescentOptimizer(0.01)

sig_train_step = opt.minimize(sig_loss)
relu_train_step = opt.minimize(relu_loss)
```

```python
init = tf.global_variables_initializer()
sess.run(init)
```

```python
epoch = 500
sig_loss_res,relu_loss_res = [],[]
sig_act,relu_act = [], []

for i in range(epoch):
    rand_index = np.random.choice(len(x), size=batch_size)
    feed_x = np.transpose([x[rand_index]])
    sess.run(sig_train_step,feed_dict={x_data:feed_x})
    sess.run(relu_train_step,feed_dict={x_data:feed_x})
    
    sig_loss_res.append(sess.run(sig_loss,feed_dict={x_data:feed_x}))
    relu_loss_res.append(sess.run(relu_loss,feed_dict={x_data:feed_x}))
    
    sig_act.append(np.mean(sess.run(sigmoid_activation,feed_dict={x_data:feed_x})))
    relu_act.append(np.mean(sess.run(relu_activation,feed_dict={x_data:feed_x})))
    
    sigmoid_output=np.mean(sess.run(sigmoid_activation,feed_dict={x_data:feed_x}))
    relu_output = np.mean(sess.run(relu_activation, feed_dict={x_data: feed_x}))
        
    if i%100==0:
        print('sigmoid = ' + str(np.mean(sigmoid_output)) + ' relu = ' + str(np.mean(relu_output)))
```

```python
plt.plot(sig_act, 'k-', label='Sigmoid Activation')
plt.plot(relu_act, 'r--', label='Relu Activation')
plt.ylim([0, 1.0])
plt.title('Activation per Generation')
plt.xlabel('Generation')
plt.ylabel('Activation')
plt.legend(loc='upper right')
```

```python
plt.plot(sig_loss_res, 'k-', label='Sigmoid loss')
plt.plot(relu_loss_res, 'r--', label='Relu loss')
plt.ylim([0, 1.0])
plt.title('Loss per Generation')
plt.xlabel('Generation')
plt.ylabel('Loss')
plt.legend(loc='upper right')
```

### 结论
sigmoid激励函数  虽然输出极值很少 但是收敛速度很慢<p>
ReLU激励函数，收敛很快，但是返回结果中容易出现极值

# 使用Tensorflow实现单层神经网络
本节将实现一个单层的神经网络，并在iris数据集上进行模型训练

```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from sklearn import datasets
from tensorflow.python.framework import ops
ops.reset_default_graph()
sess = tf.Session()
```

读取iris数据集

```python
iris = datasets.load_iris()
x =  iris.data
y = iris.target
```

```python
# 设置一个种子是的返回结果可复现
seed = 2
tf.set_random_seed(seed)
np.random.seed(seed)  
```

分割训练集和测试集 (80%/20%)

```python
train_indices = np.random.choice(len(x), round(len(x)*0.8), replace=False)
test_indices = np.array(list(set(range(len(x))) - set(train_indices)))
x_train = x[train_indices]
x_test = x[test_indices]
y_train = y[train_indices]
y_test = y[test_indices]
```

### 数据归一化: $ \frac{x_{i}-min(x)}{max(x)-min(x)}$

```python
def normalize_cols(m):
    col_max = m.max(axis=0)
    col_min = m.min(axis=0)
    return (m-col_min) / (col_max - col_min)
    
x_train = np.nan_to_num(normalize_cols(x_train))
x_test = np.nan_to_num(normalize_cols(x_test))
```

```python
epoch = 1000
batch_size = 50
hidden_layer_nodes = 10

x_data = tf.placeholder(shape=[None, 4], dtype=tf.float32)
y_target = tf.placeholder(shape=[None, 1], dtype=tf.float32)

# inputs -> hidden 
w1 = tf.Variable(tf.random_normal(shape=[4,hidden_layer_nodes])) 
b1 = tf.Variable(tf.random_normal(shape=[hidden_layer_nodes]))   
# hidden  ->  output
w2 = tf.Variable(tf.random_normal(shape=[hidden_layer_nodes,1]))
b2 = tf.Variable(tf.random_normal(shape=[1]))  
```

* 激活函数可以使用sigmoid也可以使用relu

sigmoid:<p>
hidden_output = tf.sigmoid(tf.add(tf.matmul(x_data,w1),b1))<p>
output = tf.sigmoid(tf.add(tf.matmul(hidden_output,w2),b2))

```python
# ReLU 激活函数
hidden_output = tf.nn.relu(tf.add(tf.matmul(x_data,w1),b1))
output = tf.nn.relu(tf.add(tf.matmul(hidden_output,w2),b2))
```

```python
loss = tf.reduce_mean(tf.square(tf.subtract(output,y_target)))
```

* 优化器

```python
opt = tf.train.GradientDescentOptimizer(0.005)
train_step = opt.minimize(loss)
```

```python
init = tf.global_variables_initializer()
sess.run(init)
```

```python
loss_vec = []
test_loss = []
for i in range(epoch):
    rand_index = np.random.choice(len(x_train), size=batch_size)
    rand_x = x_train[rand_index]
    rand_y = np.transpose([y_train[rand_index]])
    sess.run(train_step, feed_dict={x_data: rand_x, y_target: rand_y})

    temp_loss = sess.run(loss, feed_dict={x_data: rand_x, y_target: rand_y})
    loss_vec.append(np.sqrt(temp_loss))
    
    test_temp_loss = sess.run(loss, feed_dict={x_data: x_test, y_target: np.transpose([y_test])})
    test_loss.append(np.sqrt(test_temp_loss))
    if (i+1)%250==0:
        print('epoch: ' + str(i+1) + '.train Loss = ' + str(temp_loss)+'.test Loss = ' + str(test_temp_loss))
```

```python
plt.plot(loss_vec, 'k-', label='Train Loss')
plt.plot(test_loss, 'r--', label='Test Loss')
plt.title('Loss (MSE) per Generation')
plt.legend(loc='upper right')
plt.xlabel('Generation')
plt.ylabel('Loss')
```

# 使用TensorFlow实现神经网络的常见层
#### 本节将介绍如何实现常见的神经网络的hiddle layer（中间层）包括：
1. Convolutional Layer 卷积层
2. Activation Layer 激活层
3. Max-Pool Layer 池化层
4. Fully Connected Layer 全连接层

#                  ---------一维数据--------

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.python.framework import ops
ops.reset_default_graph()
sess = tf.Session()
```

### 定义config参数

```python
data_size = 25
conv_size = 5
maxpool_size = 5
stride_size = 1
epoch = 500

seed=13
np.random.seed(seed)
tf.set_random_seed(seed)
```

```python
data_1D = np.random.normal(size=data_size)
data_input_1D = tf.placeholder(shape=[data_size],dtype=tf.float32)
```

### 卷积层
1. 定义一个卷积层函数，
2. 声明一个随机过滤层，
3. 创建一个卷积层
4. 激活函数对卷积层操作


### 参数解释
$ OutputSize=(W-F+2P)/S+1 $
1. W为输入层维度
2. F为过滤层大小
3. P为padding的大小，即补零（在valid下，padding=0）
4. S是步长的大小

```python
def cov_layer_1d(_input,_filter):
    ''' 将一维数据转为四维数据，因为对应很多图像处理 [数量，长度，宽度，颜色通道]'''
    _input = tf.expand_dims(_input,0)
    _input = tf.expand_dims(_input,0)
    _input = tf.expand_dims(_input,3)
    
    convolution_output = tf.nn.conv2d(_input,filter=_filter,strides=[1,1,1,1],padding='VALID')
    conv_output_1d = tf.squeeze(convolution_output)
    
    return(conv_output_1d)

_filter = tf.Variable(tf.random_normal(shape=[1,conv_size,1,1]))
convolution_output = cov_layer_1d(data_input_1D,_filter)
activation_output = tf.nn.relu(convolution_output)
```

### 池化层：
1. 声明一个池化层函数
2. 该函数在一维向量的移动窗口上创建池化层函数

```python
def max_pool(_input, width,stride):
    
    _input = tf.expand_dims(_input,0)
    _input = tf.expand_dims(_input,0)
    _input = tf.expand_dims(_input,3)

    pool_output = tf.nn.max_pool(_input, ksize=[1, 1, width, 1],
                                 strides=[1, 1, stride, 1],
                                 padding='VALID')
    
    pool_output_1d = tf.squeeze(pool_output)
    return(pool_output_1d)

maxpool_output = max_pool(activation_output, width=maxpool_size,stride=stride_size)
```

### 全连接层：
创建一个函数，输入一维的数据，输出值的索引。（在一维数据做矩阵乘法时需要提取拓展到2维）

```python
def fully_connected(input_layer, outputs):

    weight_shape = tf.squeeze(tf.stack([tf.shape(input_layer),[outputs]]))
    weight = tf.random_normal(weight_shape, stddev=0.1)
    bias = tf.random_normal(shape=[outputs])
   
    input_layer_2d = tf.expand_dims(input_layer, 0)
    
    full_output = tf.add(tf.matmul(input_layer_2d, weight), bias)
    full_output_1d = tf.squeeze(full_output)
    
    return(full_output_1d)

full_output = fully_connected(maxpool_output, 5)
```

```python
init = tf.global_variables_initializer()
sess.run(init)

feed_dict = {data_input_1D: data_1D}
```

* 卷积层

```python
print('Input = array of length %d' % (data_input_1D.shape.as_list()[0]))
print('Convolution w/ filter, length = %d, stride size = %d, results in an array of length %d:' % 
      (conv_size,stride_size,convolution_output.shape.as_list()[0]))
print(sess.run(convolution_output, feed_dict=feed_dict))
```

* 激活层

```python
print('\nInput = above array of length %d' % (convolution_output.shape.as_list()[0]))
print('ReLU element wise returns an array of length %d:' % (activation_output.shape.as_list()[0]))
print(sess.run(activation_output, feed_dict=feed_dict))
```

* 池化层

```python
# Max Pool Output
print('\nInput = above array of length %d' % (activation_output.shape.as_list()[0]))
print('MaxPool, window length = %d, stride size = %d, results in the array of length %d' %
     (maxpool_size,stride_size,maxpool_output.shape.as_list()[0]))
print(sess.run(maxpool_output, feed_dict=feed_dict))
```

* 全连接层

```python
# Fully Connected Output
print('\nInput = above array of length %d' % (maxpool_output.shape.as_list()[0]))
print('Fully connected layer on all 4 rows with %d outputs:' % 
      (full_output.shape.as_list()[0]))
print(sess.run(full_output, feed_dict=feed_dict))
```

#                  --------二维数据--------

```python
# Reset Graph
ops.reset_default_graph()
sess = tf.Session()

# parameters for the run
row_size = 10
col_size = 10
conv_size = 2
conv_stride_size = 2
maxpool_size = 2
maxpool_stride_size = 1


# ensure reproducibility
seed=13
np.random.seed(seed)
tf.set_random_seed(seed)

#Generate 2D data
data_size = [row_size,col_size]
data_2d = np.random.normal(size=data_size)

x_input_2d = tf.placeholder(dtype=tf.float32, shape=data_size)

'''Convolution'''
def conv_layer_2d(input_2d, my_filter,stride_size):
   
    input_3d = tf.expand_dims(input_2d, 0)
    input_4d = tf.expand_dims(input_3d, 3)
 
    convolution_output = tf.nn.conv2d(input_4d, filter=my_filter, 
                                      strides=[1,stride_size,stride_size,1], padding="VALID")
 
    conv_output_2d = tf.squeeze(convolution_output)
    return(conv_output_2d)
'''Filter'''
my_filter = tf.Variable(tf.random_normal(shape=[conv_size,conv_size,1,1]))
my_convolution_output = conv_layer_2d(x_input_2d, my_filter,stride_size=conv_stride_size)

'''Activation'''
def activation(input_1d):
    return(tf.nn.relu(input_1d))

'''Create Activation Layer'''
my_activation_output = activation(my_convolution_output)

'''Max Pool'''
def max_pool(input_2d, width, height,stride):

    input_3d = tf.expand_dims(input_2d, 0)
    input_4d = tf.expand_dims(input_3d, 3)
    pool_output = tf.nn.max_pool(input_4d, ksize=[1, height, width, 1],
                                 strides=[1, stride, stride, 1],
                                 padding='VALID')

    pool_output_2d = tf.squeeze(pool_output)
    return(pool_output_2d)


my_maxpool_output = max_pool(my_activation_output, 
                             width=maxpool_size, height=maxpool_size,stride=maxpool_stride_size)


'''Fully Connected'''
def fully_connected(input_layer, num_outputs):
    
    flat_input = tf.reshape(input_layer, [-1])
    weight_shape = tf.squeeze(tf.stack([tf.shape(flat_input),[num_outputs]]))
    
    weight = tf.random_normal(weight_shape, stddev=0.1)
    
    bias = tf.random_normal(shape=[num_outputs])

    input_2d = tf.expand_dims(flat_input, 0)

    full_output = tf.add(tf.matmul(input_2d, weight), bias)
    
    full_output_2d = tf.squeeze(full_output)
    return(full_output_2d)

'''Create Fully Connected Layer'''
my_full_output = fully_connected(my_maxpool_output, 5)
```

```python
init = tf.global_variables_initializer()
sess.run(init)

feed_dict = {x_input_2d: data_2d}
```

```python

# Convolution Output
print('Input = %s array' % (x_input_2d.shape.as_list()))
print('%s Convolution, stride size = [%d, %d] , results in the %s array' % 
      (my_filter.get_shape().as_list()[:2],conv_stride_size,conv_stride_size,my_convolution_output.shape.as_list()))
print(sess.run(my_convolution_output, feed_dict=feed_dict))
```

```python

# Activation Output
print('\nInput = the above %s array' % (my_convolution_output.shape.as_list()))
print('ReLU element wise returns the %s array' % (my_activation_output.shape.as_list()))
print(sess.run(my_activation_output, feed_dict=feed_dict))
```

```python

# Max Pool Output
print('\nInput = the above %s array' % (my_activation_output.shape.as_list()))
print('MaxPool, stride size = [%d, %d], results in %s array' % 
      (maxpool_stride_size,maxpool_stride_size,my_maxpool_output.shape.as_list()))
print(sess.run(my_maxpool_output, feed_dict=feed_dict))
```

```python

# Fully Connected Output
print('\nInput = the above %s array' % (my_maxpool_output.shape.as_list()))
print('Fully connected layer on all %d rows results in %s outputs:' % 
      (my_maxpool_output.shape.as_list()[0],my_full_output.shape.as_list()[0]))
print(sess.run(my_full_output, feed_dict=feed_dict))
```

## 工作原理
使用神经网络算法有利有弊，在某些场景下训练更为准确，但是同时也需要很多的计算量，并且极有可能过拟合；<p>
为了解决欠拟合问题，我们可以增加训练模型的深度或者迭代的次数。<p>
为了解决过拟合问题，我们可以增加更多的数据或者使用正则化技术。
